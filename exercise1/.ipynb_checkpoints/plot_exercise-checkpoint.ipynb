```python
import numpy as np
import mesa_reader as mr
import matplotlib.pyplot as plt

import scienceplots
plt.style.use('science')

plt.rcParams.update({'font.size': 11})
```

```python
h = mr.MesaData("1Msol/LOGS/history.data")

fig, ax = plt.subplots(dpi=300, figsize=(3, 3), tight_layout=True)

Myr = 1e9 

ax.scatter(h.star_age / Myr, 10**h.log_dt / Myr, s=1) # Myr
ax.set_xlabel("Star Age [Myr]")
ax.set_ylabel(r"$\Delta t$ [Myr]")

fig.savefig("../lab_report/plots/dt_vs_age.pdf", dpi=400, bbox_inches='tight')
```

```python
# Calculate R/R_dot
```

```python
R = 10**h.log_R #in Rsun

R_dot = np.diff(R)

data_y = np.abs(R_dot / R[:-1])
```

```python
fig, ax = plt.subplots(dpi=300, figsize=(3, 3))

ax.plot(h.model_number[:-1], data_y)
ax.set_ylabel(r"$R / \dot{R}$")
ax.set_xlabel(r"Model Number")
```

```python
G = 3.94135564e-7 # Rsun^3 /Msun /s^2
M = 1 # 1Msu

t_dyn = np.sqrt(R**3 / G)
L = 10**h.log_L

t_KH = G * M**2 / (2*R*L)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, dpi=300, figsize=(3, 3), sharex=True)

ax1.loglog(h.model_number, t_dyn / 60) # hours
ax1.set_ylabel(r"$t_\text{dyn} / h$")

ax2.plot(h.model_number, t_KH) # hours
ax2.set_ylabel(r"$t_\text{KH} / h$")
ax2.set_xlabel(r"Model Number")
```
